# Trabalho Prático #1
## Treinamento de Modelos com Scikit-Learn

O objetivo deste exercício é proporcionar a estudantes de Ciência de Dados a oportunidade de praticar habilidades de treinamento e avaliação de modelos de aprendizado de máquina, utilizando o scikit-learn. Você trabalhará com três datasets populares e aplicará dois modelos para tarefas de classificação e dois para tarefas de regressão, utilizando as respectivas classes do scikit-learn.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
```

# Classificação com o Iris Dataset

```python
iris = datasets.load_iris()

scaler = StandardScaler()

iris_scaled = scaler.fit_transform(iris.data)

X_train, X_test, y_train, y_test = train_test_split(iris_scaled, iris.target, test_size=0.9, random_state=42)

iris_knn_classifier = KNeighborsClassifier()
iris_knn_classifier.fit(X_train, y_train)
y_pred_knn = iris_knn_classifier.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)

iris_svc_model = svm.SVC()
iris_svc_model.fit(X_train, y_train)
y_pred_svc = iris_svc_model.predict(X_test)
accuracy_svc = accuracy_score(y_test, y_pred_svc)

iris_LR_model = LinearRegression()
iris_LR_model.fit(X_train, y_train)
y_pred_lr = iris_LR_model.predict(X_test)
r2_lr = r2_score(y_test, y_pred_lr)

iris_RFR_model = RandomForestRegressor()
iris_RFR_model.fit(X_train, y_train)
y_pred_rfr = iris_RFR_model.predict(X_test)
r2_rfr = r2_score(y_test, y_pred_rfr)

print("KNeighborsClassifier accuracy: ", end=" ")
print(accuracy_knn)
print("SVC accuracy: ", end=" ")
print(accuracy_svc)
print("Linear Regression - Coefficient of Determination (R²): ", end=" ")
print(r2_lr)
print("Random Forest Regression - Coefficient of Determination (R²): ", end=" ")
print(r2_rfr)

```

```python
# Converter os dados em um DataFrame do pandas
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Adicionar a coluna de target ao DataFrame
iris_df['target'] = iris.target

print(iris.target_names)

# Plotar um gráfico de dispersão usando matplotlib
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=iris_df['target'], cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset - Sepal Length vs Sepal Width')
plt.colorbar(label='Species: 0 - Setosa, 1 - versicolor, 2 - virginica')
plt.show()
```

```python
wines = datasets.load_wine()

scaler = StandardScaler()

wines_scaled = scaler.fit_transform(wines.data)

X_train, X_test, y_train, y_test = train_test_split(wines_scaled, wines.target, test_size=0.9, random_state=42)

wines_knn_classifier = KNeighborsClassifier()
wines_knn_classifier.fit(X_train, y_train)
y_pred_knn = wines_knn_classifier.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)

wines_svc_model = svm.SVC()
wines_svc_model.fit(X_train, y_train)
y_pred_svc = wines_svc_model.predict(X_test)
accuracy_svc = accuracy_score(y_test, y_pred_svc)

wines_LR_model = LinearRegression()
wines_LR_model.fit(X_train, y_train)
y_pred_lr = wines_LR_model.predict(X_test)
r2_lr = r2_score(y_test, y_pred_lr)

wines_RFR_model = RandomForestRegressor()
wines_RFR_model.fit(X_train, y_train)
y_pred_rfr = wines_RFR_model.predict(X_test)
r2_rfr = r2_score(y_test, y_pred_rfr)

print("KNeighborsClassifier accuracy: ", end=" ")
print(accuracy_knn)
print("SVC accuracy: ", end=" ")
print(accuracy_svc)
print("Linear Regression - Coefficient of Determination (R²): ", end=" ")
print(r2_lr)
print("Random Forest Regression - Coefficient of Determination (R²): ", end=" ")
print(r2_rfr)

```
